2. What is the difference between string and variable?

Ans: String is a sequence of characters. Strings are always enclosed within ‘’ or “” i.e. single quotes or double quotes. A number can also be converted into string by using str or putting it into ‘’ or “”. For example: “iNeuron”, ‘a’
Whereas variable is an alphabet or set of alphabets to whom we can assign any values, For example: a = 7,  abc = 8. Here a & abc are two variables which have been assigned a value. The assignment of a variable can be changed. 


3. Describe three different data types.

Ans: The three data types are integer, float, complex number.
Integer: Whole numbers along with negative of natural numbers constitute integers. For ex: -1,0,4
Float: Decimal numbers are known as float. For ex: 1.1, 2.3 etc
Complex numbers: A number in the form a + bj is known as complex number. Here j^2 = -1!. For ex: 4+6j
